```python
import numpy as np
import pandas as pd

import warnings
warnings.filterwarnings('ignore')
```

```python
train = pd.read_csv('../input/netflix-appetency/train.csv')
test = pd.read_csv('../input/netflix-appetency/test.csv')
```

```python
from autogluon.tabular import TabularPredictor
```

```python
label = 'target'
eval_metric = 'roc_auc'
save_path = '/AutoGluons/'

predictor = TabularPredictor(label=label, eval_metric=eval_metric, 
                             path=save_path, verbosity=2)
predictor.fit(train, presets='best_quality', time_limit=1200)

results = predictor.fit_summary()
```

```python
preds = predictor.predict_proba(test)[1]
```

```python
submission = pd.DataFrame({'id': test.id, 'target': preds})
```

```python
submission.to_csv('submission.csv', index=False)
!head submission.csv
```
